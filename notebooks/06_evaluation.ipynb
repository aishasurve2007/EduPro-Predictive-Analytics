```python
import pandas as pd

enrollment_results = pd.read_csv(
    "../data/enrollment_results.csv"
)

revenue_results = pd.read_csv(
    "../data/revenue_results.csv"
)

print("Enrollment Models")
display(enrollment_results.sort_values("R2", ascending=False))

print("Revenue Models")
display(revenue_results.sort_values("R2", ascending=False))
```

```python
import matplotlib.pyplot as plt

enrollment_results.sort_values(
    "R2",
    ascending=False
).plot(
    x="Model",
    y="R2",
    kind="bar"
)

plt.title("Enrollment Model Comparison")
plt.ylabel("R² Score")
plt.show()
```

```python
revenue_results.sort_values(
    "R2",
    ascending=False
).plot(
    x="Model",
    y="R2",
    kind="bar"
)

plt.title("Revenue Model Comparison")
plt.ylabel("R² Score")
plt.show()
```
