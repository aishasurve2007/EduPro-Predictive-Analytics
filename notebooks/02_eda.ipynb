```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/model_dataset.csv")

print(df.shape)
df.head()
```

```python
df.info()

df.describe(include="all")
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["EnrollmentCount"], bins=20)
plt.title("Enrollment Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["TotalRevenue"], bins=20)
plt.title("Revenue Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["CoursePrice"], bins=20)
plt.title("Course Price Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["CourseRating"], bins=20)
plt.title("Course Rating Distribution")
plt.show()
```

```python
category_revenue = (
    df.groupby("CourseCategory")["TotalRevenue"]
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(10,5))
category_revenue.plot(kind="bar")
plt.title("Revenue by Category")
plt.ylabel("Revenue")
plt.show()
```

```python
category_enrollment = (
    df.groupby("CourseCategory")["EnrollmentCount"]
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(10,5))
category_enrollment.plot(kind="bar")
plt.title("Enrollment by Category")
plt.show()
```

```python
numeric_cols = [
    "CoursePrice",
    "CourseDuration",
    "CourseRating",
    "YearsOfExperience",
    "TeacherRating",
    "EnrollmentCount",
    "TotalRevenue",
    "AvgRevenue",
    "UniqueStudents"
]
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    df[numeric_cols].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")
plt.show()
```
