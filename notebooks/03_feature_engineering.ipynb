```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/model_dataset.csv")

print(df.shape)
df.head()
```

```python
df.info()
```

```python
df["PriceBand"] = pd.cut(
    df["CoursePrice"], 
    bins=3, 
    labels=["Low", "Medium", "High"]
)
```

```python
df["DurationBucket"] = pd.cut(
    df["CourseDuration"],
    bins=[0, 10, 20, 100],
    labels=["Short", "Medium", "Long"]
)

df[["CourseDuration", "DurationBucket"]].head()
```

```python
df["RatingTier"] = pd.cut(
    df["CourseRating"],
    bins=[0, 3.5, 4.5, 5],
    labels=["Low", "Medium", "High"]
)

df[["CourseRating", "RatingTier"]].head()
```

```python
# 1. Check the actual data distribution of the raw column
print("Summary Statistics of Experience:")
print(df["YearsOfExperience"].describe())

# 2. Look at the raw numbers directly
print("\nFirst 10 raw values:")
print(df["YearsOfExperience"].head(10).values)

# 3. Check if there are any values actually below 10
print("\nHow many rows are less than or equal to 5?")
print((df["YearsOfExperience"] <= 5).sum())
```

```python
df["ExperienceBucket"] = pd.cut(
    df["YearsOfExperience"],
    bins=[0, 5, 10, 50],
    labels=["Junior", "Mid", "Senior"]
)

df[
    ["YearsOfExperience", "ExperienceBucket"]
].head()
```

```python
df["ExpertiseMatch"] = (
    df["CourseCategory"].str.lower()
    ==
    df["Expertise"].str.lower()
).astype(int)

df[
    ["CourseCategory",
     "Expertise",
     "ExpertiseMatch"]
].head()
```

```python
df["RevenuePerEnrollment"] = (
    df["TotalRevenue"]
    /
    df["EnrollmentCount"]
)

df[
    ["TotalRevenue",
     "EnrollmentCount",
     "RevenuePerEnrollment"]
].head()
```

```python
level_map = {
    "Beginner": 1,
    "Intermediate": 2,
    "Advanced": 3
}

df["LevelEncoded"] = (
    df["CourseLevel"]
    .map(level_map)
)

df[
    ["CourseLevel",
     "LevelEncoded"]
].head()
```

```python
df["PopularityScore"] = (
    df["EnrollmentCount"]
    *
    df["CourseRating"]
)

df[
    ["EnrollmentCount",
     "CourseRating",
     "PopularityScore"]
].head()
```

```python
df["RevenueTier"] = pd.cut(
    df["TotalRevenue"], 
    bins=3, 
    labels=["Low", "Medium", "High"]
)
```

```python
new_features = [
    "PriceBand",
    "DurationBucket",
    "RatingTier",
    "ExperienceBucket",
    "ExpertiseMatch",
    "RevenuePerEnrollment",
    "LevelEncoded",
    "PopularityScore",
    "RevenueTier"
]

df[new_features].head()
```

```python
df.isnull().sum()
```

```python
df.to_csv(
    "../data/feature_engineered_dataset.csv",
    index=False
)

print(
    "feature_engineered_dataset.csv saved successfully"
)
```
